## 文心开源模型的选择

本项目聚焦企业级发票处理场景的本地化部署需求，在模型选型阶段对ERNIE-4.5系列模型进行了全面评估，最终选择**ERNIE-4.5-21B-A3B-Paddle**作为核心大模型，配合PP-OCR实现全流程本地化部署。

### 选型理由
本项目选择**ERNIE-4.5-21B-A3B-Paddle**的核心原因：  
1. **性能适配**：21B总参数/3B激活参数的规模，既能满足发票信息抽取、结构化解析等专业任务的精度需求，又避免了超大规模模型的部署复杂度；  
2. **部署可行性**：支持单机多卡部署（4×V100即可满足），无需GPU集群，降低企业本地化部署的硬件门槛；  
3. **上下文能力**：128K超长上下文窗口，可完整处理多页发票扫描件的文本解析；  
4. **Paddle生态兼容**：与PP-OCR同属飞桨生态，模型调用与协同效率更高，且支持FastDeploy快速部署。  


## 项目背景

在企业财务数字化转型过程中，传统发票处理流程面临人工识别效率低、数据录入易错、信息提取不完整等挑战，且外部API调用存在数据隐私泄露风险。本项目基于飞桨开源框架的**PP-OCR**与百度**ERNIE-4.5-21B-A3B**大模型，构建了一套**全本地化部署**的智能发票处理系统，通过OCR技术与大模型的协同，实现发票信息的自动化提取、结构化存储与风险分析，全程数据不落地外部服务器，满足企业数据安全合规要求。


## 项目立意

- **全栈本地化部署**：PP-OCR与ERNIE-4.5-21B-A3B模型均部署在企业内部服务器，数据处理全程闭环，保障财务数据隐私；  
- **智能识别融合**：整合PP-OCR的高精度文字识别与ERNIE-4.5的深度语义理解能力，实现发票关键信息（如价税分离、购销方识别）的精准提取；  
- **流程自动化**：通过Web界面实现"OCR识别→企业信息查询→风险分析"的一站式处理，支持单张/批量发票处理，大幅提升财务工作效率；  
- **低门槛运维**：基于FastDeploy实现模型一键部署，支持7000端口本地服务调用，降低企业技术团队的运维成本。

def extract_text_from_image(self, img_path: str) -> str:
    result = self.ocr.predict(img_path)  # 本地推理，无外部依赖
    text = '\n'.join([t.strip() for t in result[0]["rec_texts"] if t.strip()])
    return text

def process_invoice_step(image, step, current_state=None):
    """分步处理发票（无外部API参数，仅调用本地服务）"""
    if step == "ocr":
        # 第一步：本地PP-OCR识别
        result = processor.process_invoice_basic(image)  # 调用本地OCR
    elif step == "company_info":
        # 第二步：本地企业信息查询

#### 3.2 批量处理自动化（本地Excel生成）
```python
def process_multiple_invoices(self, image_paths: List[str]) -> Dict[str, Any]:
    """批量处理并生成本地Excel报告"""
    results = []
    for image_path in image_paths:
        result = self.process_invoice_basic(image_path)  # 本地OCR+信息提取
        company_info = self.get_company_information(result)  # 本地企业信息
        analysis = self.generate_analysis_report(result, company_info)  # 本地大模型分析
        results.append({"basic_info": result, "company_info": company_info, "analysis": analysis})
    # 结果保存至本地Excel（不依赖云端存储）
    excel_path = os.path.join("output", f"发票处理结果_{timestamp}.xlsx")
    with pd.ExcelWriter(excel_path) as writer:
        pd.DataFrame(results).to_excel(writer)
    return {"excel_path": excel_path}
```
